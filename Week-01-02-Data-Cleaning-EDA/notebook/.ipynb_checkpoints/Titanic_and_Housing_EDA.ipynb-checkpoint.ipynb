### Dataset 1: Titanic

#### 1. Dataset Loading and Inspection

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load Titanic dataset
df = pd.read_csv(r"C:\Users\RUTH\OneDrive\Desktop\AnalystLab Africa Internship\Titanic-Dataset.csv")
# first 10 rows
df.head(10)
```

```python
# Number of rows and columns
df.shape
```

```python
# Summary Statistics
df.info()
```

```python
df.describe()
```

```python
#Check for missing values
df.isnull().sum()
```

#### 2. Handling Missing Values
##### i. Age Column - Grouped Median

```python
# Extract titles from names
df["Title"] = df["Name"].str.extract(r',\s*([^\.]+)\.')
```

```python
df["Title"].unique()
```

```python
df['Title'].value_counts()
```

```python
# Handle rare titles
df["Title"] = df["Title"].replace({
    "Mlle": "Miss",
    "Ms": "Miss",
    "Mme": "Mrs",
    "Lady": "Rare",
    "the Countess": "Rare",
    "Capt": "Rare",
    "Col": "Rare",
    "Don": "Rare",
    "Dr": "Rare",
    "Major": "Rare",
    "Rev": "Rare",
    "Sir": "Rare",
    "Jonkheer": "Rare",
    })
```

```python
df["Title"].unique()
```

```python
# Group by Title and Class
df["Age"] = df["Age"].fillna(
    df.groupby(["Title", "Pclass"])["Age"].transform("median")
)
df = df.drop(columns = 'Title')
```

```python
df['Age'].isnull().sum()
```

#### ii. Cabin Column (Extract deck and drop the column)

```python
df['Deck'] = df['Cabin'].str[0].fillna('Unknown')
df = df.drop(columns = 'Cabin')
```

#### iii. Embarked Column

```python
df['Embarked'] = df['Embarked'].fillna('C')
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
# Save cleaned dataset
df.to_csv('titanic_cleaned.csv', index=False)

print("Cleaned Titanic dataset saved successfully!")
```

# Univariate and Multivariate Analysis
## 1. Univariate Analysis

```python
# Distribution of Numerical Features
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()  # Flatten the 2D array to 1D for easy looping

for i, col in enumerate(numerical_cols):
    sns.histplot(df[col], kde=True, ax=axes[i], bins=20)
    axes[i].set_title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

```python
# Detect Outliers
plt.figure(figsize=(10,6))

for i, column in enumerate(numerical_cols):
    plt.subplot(2,2,i+1)
    sns.boxplot(y=df[column])
    plt.title(column)

plt.tight_layout()
plt.show()
```

```python
# Categorical Features
categorical_cols = ['Sex', 'Pclass', 'Embarked', 'Survived']

# grid layout
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

# 3. Loop through and create a countplot for each feature
for i, col in enumerate(categorical_cols):
    # 'hue' is set to the column itself to automatically give each bar a distinct color
    sns.countplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f'Frequency Distribution of {col}')
    axes[i].set_ylabel('Count')
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()
```

## 2. Multivariate Analysis

```python
# Categorical Features
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 1. Survival by Gender (Countplot)
sns.countplot(data=df, x='Sex', hue='Survived', ax=axes[0,0])
axes[0,0].set_title('Survival by Gender')
axes[0,0].set_xlabel('Gender')
axes[0,0].set_ylabel('Count')
axes[0,0].legend(title='Survived', labels=['No', 'Yes'])

# 2. Average Survival Rate by Gender
sns.barplot(data=df, x='Sex', y='Survived', ax=axes[0,1])
axes[0,1].set_title('Average Survival Rate by Gender')
axes[0,1].set_xlabel('Gender')
axes[0,1].set_ylabel('Survival Rate')

# 3. Survival by Passenger Class
sns.countplot(data=df, x='Pclass', hue='Survived', ax=axes[0,2])
axes[0,2].set_title('Survival by Passenger Class')
axes[0,2].set_xlabel('Passenger Class')
axes[0,2].set_ylabel('Count')
axes[0,2].legend(title='Survived', labels=['No', 'Yes'])

# 4. Average Survival Rate by Passenger Class
sns.barplot(data=df, x='Pclass', y='Survived', ax=axes[1,0])
axes[1,0].set_title('Average Survival Rate by Passenger Class')
axes[1,0].set_xlabel('Passenger Class')
axes[1,0].set_ylabel('Survival Rate')

# 5. Survival by Embarkation Port
sns.countplot(data=df, x='Embarked', hue='Survived', ax=axes[1,1])
axes[1,1].set_title('Survival by Embarkation Port')
axes[1,1].set_xlabel('Embarkation Port')
axes[1,1].set_ylabel('Count')
axes[1,1].legend(title='Survived', labels=['No', 'Yes'])

# 6. Survival Rate by Number of Siblings/Spouses
sns.barplot(data=df, x='SibSp', y='Survived', ax=axes[1,2])
axes[1,2].set_title('Survival Rate by Number of Siblings/Spouses')
axes[1,2].set_xlabel('Number of Siblings/Spouses')
axes[1,2].set_ylabel('Survival Rate')

plt.tight_layout()
plt.show()
```

```python
# Numerical Features
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Age vs Survival
sns.boxplot(data=df, x='Survived', y='Age', ax=axes[0,0])
axes[0,0].set_title('Age vs Survival')

# Fare vs Survival
sns.boxplot(data=df, x='Survived', y='Fare', ax=axes[0,1])
axes[0,1].set_title('Fare vs Survival')

# Survival Rate by Passenger Class
sns.barplot(data=df, x='Pclass', y='Survived', ax=axes[1,0])
axes[1,0].set_title('Average Survival Rate by Passenger Class')

# Survival Rate by Parents/Children
sns.barplot(data=df, x='Parch', y='Survived', ax=axes[1,1])
axes[1,1].set_title('Survival Rate by Parents/Children')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Age',
    y='Fare',
    hue='Survived'
)

plt.title('Age vs Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))

corr = df[['Survived','Pclass','Age','SibSp','Parch','Fare']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")

plt.title("Correlation Matrix")

plt.savefig('titanic_correlation_matrix.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Age vs Passenger Class
sns.boxplot(data=df, x='Pclass', y='Age', ax=axes[0])
axes[0].set_title('Age Distribution by Passenger Class')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')

# Fare vs Passenger Class
sns.boxplot(data=df, x='Pclass', y='Fare', ax=axes[1])
axes[1].set_title('Fare Distribution by Passenger Class')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')

plt.tight_layout()
plt.show()
```

## EDA - Univariate and Bivariate Analysis
### 1. Univariate Analysis

```python
# Numerical Features
numerical_cols = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']

housing_df[numerical_cols].hist(figsize=(15,10), bins=20)

plt.suptitle('Distribution of Numerical Variables')
plt.show()
```

```python
# Outlier Detection
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

for ax, column in zip(axes.flatten(), numerical_cols):
    sns.boxplot(y=housing_df[column], ax=ax)
    ax.set_title(f'{column.capitalize()}')

plt.tight_layout()
plt.show()
```

```python
# Categorical Features
categorical_cols = [
    'mainroad',
    'guestroom',
    'basement',
    'hotwaterheating',
    'airconditioning',
    'prefarea',
    'furnishingstatus'
]

fig, axes = plt.subplots(3, 3, figsize=(15,12))

for ax, column in zip(axes.flatten(), categorical_cols):
    sns.countplot(data=housing_df, x=column, ax=ax)
    ax.set_title(column.capitalize())

# Hide unused subplots
for ax in axes.flatten()[len(categorical_cols):]:
    ax.set_visible(False)

plt.tight_layout()
plt.show()
```

### 2. Bivariate Analysis

```python
## Numerical Features

fig, axes = plt.subplots(2, 2, figsize=(14,10))

# Price vs Area
sns.scatterplot(data=housing_df, x='area', y='price', ax=axes[0,0])
axes[0,0].set_title('Price vs Area')

# Price vs Bedrooms
sns.boxplot(data=housing_df, x='bedrooms', y='price', ax=axes[0,1])
axes[0,1].set_title('Price vs Bedrooms')

# Price vs Bathrooms
sns.boxplot(data=housing_df, x='bathrooms', y='price', ax=axes[1,0])
axes[1,0].set_title('Price vs Bathrooms')

# Price vs Stories
sns.boxplot(data=housing_df, x='stories', y='price', ax=axes[1,1])
axes[1,1].set_title('Price vs Stories')

plt.tight_layout()
plt.show()
```

```python
## Categorical Features

fig, axes = plt.subplots(2, 3, figsize=(16,10))

sns.boxplot(data=housing_df, x='mainroad', y='price', ax=axes[0,0])
axes[0,0].set_title('Price vs Main Road')

sns.boxplot(data=housing_df, x='guestroom', y='price', ax=axes[0,1])
axes[0,1].set_title('Price vs Guestroom')

sns.boxplot(data=housing_df, x='basement', y='price', ax=axes[0,2])
axes[0,2].set_title('Price vs Basement')

sns.boxplot(data=housing_df, x='airconditioning', y='price', ax=axes[1,0])
axes[1,0].set_title('Price vs Air Conditioning')

sns.boxplot(data=housing_df, x='prefarea', y='price', ax=axes[1,1])
axes[1,1].set_title('Price vs Preferred Area')

sns.boxplot(data=housing_df, x='furnishingstatus', y='price', ax=axes[1,2])
axes[1,2].set_title('Price vs Furnishing Status')

plt.tight_layout()
plt.show()
```

```python
# Encode Categorical Features for Correlation Analysis
house_encoded = housing_df.copy()

binary_cols = [
    'mainroad',
    'guestroom',
    'basement',
    'hotwaterheating',
    'airconditioning',
    'prefarea'
]

for col in binary_cols:
    house_encoded[col] = house_encoded[col].map({'yes':1,'no':0})

house_encoded = pd.get_dummies(
    house_encoded,
    columns=['furnishingstatus'],
    drop_first=True
)
```

```python
## Correlation Heatmap
plt.figure(figsize=(12,8))

corr = house_encoded.corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Matrix')
plt.savefig('housing_correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Save Cleaned Housing Dataset
housing_df.to_csv("Cleaned_Housing_Dataset", index= False)
print("Cleaned Housing dataset saved successfully!")
```
